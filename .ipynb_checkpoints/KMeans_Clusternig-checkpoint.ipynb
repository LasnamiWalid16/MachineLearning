<h2>Clustering Kmeans Method</h2>


```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import sklearn
from sklearn.preprocessing import scale
import sklearn.metrics as sm
from sklearn.metrics import confusion_matrix, classification_report

from sklearn.cluster import KMeans
from mpl_toolkits.mplot3d import Axes3D
from sklearn import datasets
```

```python
%matplotlib inline
plt.figure(figsize=(7,4))
```

<h3>Importing IRIS data_set</h3>

```python
iris = datasets.load_iris()
X = scale(iris.data)
y = pd.DataFrame(iris.target)
variable_names = iris.feature_names
```

<h3>Fiting our MODEL KMEANS</h3>

```python
clustering = KMeans(n_clusters=3,random_state=5)
clustering.fit(X)
```

<h3>Ploting our Model to see the deff Groups =3</h3>

```python
iris_df = pd.DataFrame(iris.data)
iris_df.columns=['Sepal_Length','Sepal_width','Petal_Length','Petal_Width']
y.columns=['Targets']

```

```python
color_theme=np.array(['darkgray','lightsalmon','powderblue'])

plt.subplot(1,2,1)
plt.scatter(x=iris_df.Petal_Length,y=iris_df.Petal_Width,c=color_theme[iris.target],s=50)
plt.title('Ground Truth Classification')

plt.subplot(1,2,2)
plt.scatter(x=iris_df.Petal_Length,y=iris_df.Petal_Width,c=color_theme[clustering.labels_],s=50)
plt.title('KMeans Classification')
```

```python
relabel = np.choose(clustering.labels_,[2,0,1]).astype(np.int64)
plt.subplot(1,2,1)
plt.scatter(x=iris_df.Petal_Length,y=iris_df.Petal_Width,c=color_theme[iris.target],s=50)
plt.title('Ground Truth Classification_Petal')

plt.subplot(1,2,2)
plt.scatter(x=iris_df.Petal_Length,y=iris_df.Petal_Width,c=color_theme[relabel],s=50)
plt.title('KMeans Classification_Petal')
```

```python
relabel = np.choose(clustering.labels_,[2,0,1]).astype(np.int64)
plt.subplot(1,2,1)
plt.scatter(x=iris_df.Sepal_Length,y=iris_df.Sepal_width,c=color_theme[iris.target],s=50)
plt.title('Ground Truth Classification_Sepal')

plt.subplot(1,2,2)
plt.scatter(x=iris_df.Sepal_Length,y=iris_df.Sepal_width,c=color_theme[relabel],s=50)
plt.title('KMeans Classification_Sepal')
```
